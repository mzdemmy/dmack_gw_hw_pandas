```python
# Dependencies
import pandas as pd
```

```python
# load CSV
csv_path = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
new_coders_df = pd.read_csv(csv_path, encoding = 'iso-8859-1', low_memory=False)
```

```python
# Inspect all columns
list(new_coders_df.columns)
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
re_newcoders_df = new_coders_df.iloc[:,[0,1,2,3,4,7,8,9,10,11,29,30,32,36,37,45,48,56,110,111]]
re_newcoders_df.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
re_newcoders_df = re_newcoders_df.replace({0.0: "No", 1.0: "Yes"})
re_newcoders_df.head()
```

```python
# Calculate total number of respondents in survey
tot_responses = len(re_newcoders_df)
tot_responses
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attended_bootcamp = re_newcoders_df.loc[re_newcoders_df["AttendedBootcamp"]=="Yes"]
attended_bootcamp.head()
```

```python
# Calculate average age of attendees
avg_age = attended_bootcamp["Age"].mean()

# Calculate how many people attended a bootcamp
bootcamp_ct = attended_bootcamp["AttendedBootcamp"].count()

# Calculate how many attendees hold degrees
hold_degrees = attended_bootcamp["SchoolDegree"].count()

# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
tot_genderid = attended_bootcamp["Gender"].count()
female = attended_bootcamp["Gender"].value_counts()["female"]
male = attended_bootcamp["Gender"].value_counts()["male"]
nonbinary = tot_genderid - female - male

# Calculate percentage of respondents who attended a bootcamp
pct_bootcamp = (bootcamp_ct / tot_responses)*100

# Calculate percentage of respondents belonging to each gender
pct_female = (female / tot_genderid)*100
pct_male = (male / tot_genderid)*100
pct_nonbinary = (nonbinary / tot_genderid)*100

# Calculate percentage of attendees with a school degree
pct_degree = (hold_degrees / tot_responses)*100

# Calculate average post-bootcamp salary
avg_salary = attended_bootcamp["BootcampPostSalary"].mean()
avg_salary

#display variables
print(avg_age)
print(bootcamp_ct)
print(hold_degrees)
print(tot_genderid)
print(female)
print(nonbinary)
print(pct_bootcamp)
print(pct_female)
print(pct_male)
print(pct_nonbinary)
print(pct_degree)
print(avg_salary)
```

```python
# Create a new table consolodating above calculations
bootcamp_summary = pd.DataFrame({"Total Surveyed": [tot_responses],
                               "Total Attendees": [bootcamp_ct],
                               "% Attended Bootcamp": [pct_bootcamp],
                               "Average Age": [avg_age],
                               "% Male": [pct_male],
                               "% Female": [pct_female],
                               "% Non-Binary": [pct_nonbinary],
                               "% Who Earned Degree": [pct_degree],
                               "Average Post Bootcamp Salary": [avg_salary]
})
bootcamp_summary = bootcamp_summary[["Total Surveyed", "Total Attendees", "% Attended Bootcamp","Average Age","% Male", 
                                "% Female","% Non-Binary","% Who Earned Degree","Average Post Bootcamp Salary"]]
bootcamp_summary = bootcamp_summary.round(2)

bootcamp_summary
```

```python
# Improve formatting before outputting spreadsheet
bootcamp_summary["% Attended Bootcamp"]=bootcamp_summary["% Attended Bootcamp"].map("{0:,.2f}%".format)
bootcamp_summary["% Male"]=bootcamp_summary["% Male"].map("{0:,.2f}%".format)
bootcamp_summary["% Female"]=bootcamp_summary["% Female"].map("{0:,.2f}%".format)
bootcamp_summary["% Non-Binary"]=bootcamp_summary["% Non-Binary"].map("{0:,.2f}%".format)
bootcamp_summary["% Who Earned Degree"]=bootcamp_summary["% Who Earned Degree"].map("{0:,.2f}%".format)
bootcamp_summary["Average Post Bootcamp Salary"]=bootcamp_summary["Average Post Bootcamp Salary"].map("${0:,.2f}".format)

bootcamp_summary
```

```python
# Export to Excel
bootcamp_summary.to_excel("output/Task1A.xlsx", index=False)
```
